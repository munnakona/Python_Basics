# 1) EDA Technique:  Missing Value Treatment

```python
clean_data['Age']
```

```python
import numpy as np
```

```python
clean_data['Age'] = clean_data['Age'].fillna(np.mean(pd.to_numeric(clean_data['Age']))) # Apply Mean to the Missing Data using Numpy for Numerical Value
```

```python
clean_data
```

```python
clean_data['Exp'] = clean_data['Exp'].fillna(np.mean(pd.to_numeric(clean_data['Exp'])))
```

```python
clean_data
```

```python
clean_data['Location']
```

```python
clean_data['Location'] = clean_data['Location'].fillna(clean_data['Location'].mode()[0]) # This is applied Mode for Categorical value
```

```python
clean_data['Location']
```

```python
clean_data 
```

```python
clean_data
```

```python
clean_data['Age'] = clean_data['Age'].astype(int) # Change numerical data type to int
```

```python
clean_data['Exp'] = clean_data['Exp'].astype(int)
```

```python
clean_data['Salary'] = clean_data['Salary'].astype(int)
```

```python
clean_data
```

```python
clean_data.info()
```
